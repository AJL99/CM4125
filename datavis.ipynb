# Nobel Prize Winners data import and trimming

```python
NP = pd.read_csv('/content/my data/nobel_prize_by_winner.csv', encoding='latin-1')
```

```python
NP
```

```python
NP_trimmed = NP[['firstname', 'surname', 'bornCountry']]
```

```python
NP_trimmed.dropna(inplace=True)
```

```python
NP_trimmed.rename(columns={'bornCountry': 'country'}, inplace=True)
```

```python
NP_trimmed.replace('USA', 'United States', inplace=True)
```

```python
NP_trimmed
```

```python
for NPfix, row in NP_trimmed.iterrows():
  if NP_trimmed.loc[NPfix].at['country'].find('\('):
    NP_trimmed['country'] = NP_trimmed['country'].str.replace(r'^.*\(', '').str.slice(4, -1)
```

```python
for NPfix, row in NP_trimmed.iterrows():
    if NP_trimmed.loc[NPfix].at['country'].find('(') >= 0:
        split_result = NP_trimmed.loc[NPfix, 'country'].split('(now ')
        if len(split_result) >= 2:
            NP_trimmed.loc[NPfix, 'country'] = split_result[1][:-1]
```

```python
NP_trimmed
```

```python
NP_trimmed.to_csv('/content/my data/trimmed/NP_trimmed.csv', index=False)
```

# Table merging

```python
filteredCSV = pd.merge(DR_trimmed, IQ_trimmed, on='country', how='inner')
filteredCSV = pd.merge(filteredCSV, MS_trimmed, on='country', how='inner')
filteredCSV = pd.merge(filteredCSV, NP_trimmed, on='country', how='inner')
filteredCSV
```

```python
filteredCSV.to_csv('/content/my data/trimmed/filtered.csv')
# first merge
```

```python
country_count = filteredCSV.groupby('country').size().reset_index(name='number of nobel prize winners')
```

```python
CountedNobels = pd.merge(filteredCSV, country_count, on='country')
```

```python
CountedNobels
```

```python
CountedNobelsTrimmed = CountedNobels.drop_duplicates(subset=['country'])
CountedNobelsTrimmed.drop(['firstname', 'surname'], axis=1, inplace=True)
```

```python
CountedNobelsTrimmed
```

```python
CountedNobelsTrimmed.to_csv('/content/my data/trimmed/WorkingData.csv', index=False)
```
